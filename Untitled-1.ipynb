```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import word_tokenize
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
url = "https://www.urvann.com/product/aparajita-blue-in-4-inch-nursery-pot"

response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.content , 'html.parser')
soup
```

```python
data = []

for articles in soup.find_all('div' , class_= "ms-mt-xs ms-mb-s product-name ng-scope"):
    headline = articles.find()
    if headline:
        headline = headline.text.strip('div')
    else:
        continue
    
    data.append({'headline': headline})
```
